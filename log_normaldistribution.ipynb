```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(
    "income.csv",
    index_col=None,
    names=["income","count"],
    skiprows=1
)
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(),
                          rotation=45,
                          horizontalalignment='right');
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(),
                          rotation=45,
                          horizontalalignment='right');
g.set(xscale="log");
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = pd.DataFrame({
    "company": ["A", "B", "C", "D", "E"],
    "revenue_mln": [50, 70, 40, 90, 60]
})

# Define custom colors (named or HEX codes)
colors = ["#FF5733", "#33FF57", "#3357FF", "#FF33A8", "#FFC300"]

# Plot with different colors for each bar
df.plot(x="company", y="revenue_mln", kind="bar", color=colors, legend=False)

plt.title("Company Revenue (Colorful Bar Chart)")
plt.xlabel("Company")
plt.ylabel("Revenue (mln)")
plt.show()
```
